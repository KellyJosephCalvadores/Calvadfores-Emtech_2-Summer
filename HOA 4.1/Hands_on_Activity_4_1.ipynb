##Part 3: Perform Simple Linear Regression on the SURVIVAL feature column (you can check the internet on how you can perform simple linear regression)

```python
X = TrainData[['Age']]
y = TrainData['Survived']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=123)
```

```python
Model = LinearRegression()
Model.fit(X_train, y_train)
```

```python
YpredTrain = Model.predict(X_train)
YpredTest = Model.predict(X_test)
```

```python
mse_train = mean_squared_error(y_train, YpredTrain)*100
mse_test = mean_squared_error(y_test, YpredTest)*100
r2_train = r2_score(y_train, YpredTrain)*100
r2_test = r2_score(y_test, YpredTest)*100

print(f'Train Mean Squared Error: {mse_train}')
print(f'Test Mean Squared Error: {mse_test}')
print(f'Train R^2 Score: {r2_train}')
print(f'Test R^2 Score: {r2_test}')
```

```python
plt.subplot(1, 2, 1)
plt.scatter(X_train, y_train, color='blue', label='Actual')
plt.plot(X_train, YpredTrain, color='red', linewidth=2, label='Predicted')
plt.title('Train Set: Survival Rate')
plt.legend()

plt.subplot(1, 2, 2)
plt.scatter(X_test, y_test, color='blue', label='Actual')
plt.plot(X_test, YpredTest, color='red', linewidth=2, label='Predicted')
plt.title('Test Set: Age vs Survived')
plt.legend()

plt.show()
```

###Step 3: Train and Score the Decision Tree Model.

**a) Create an array object with the variable that will be the target for the model.**

```python
y_target = Training['Survived'].values
```

**b) Create an array of the values that will be the input for the model**

```python
columns = ["Fare", "Pclass", "Sex", "Age", "SibSp"]
X_input = Training[list(columns)].values
```
