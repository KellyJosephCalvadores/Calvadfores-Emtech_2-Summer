```python
import pandas as pd
from keras.models import Sequential
from keras.layers import Dense
from scikeras.wrappers import KerasClassifier
from tensorflow.keras.utils import to_categorical
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import KFold
from sklearn.preprocessing import LabelEncoder
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

Resource: https://archive.ics.uci.edu/dataset/186/wine+quality

```python
CData = pd.read_csv('/content/drive/MyDrive/CPE 019 (Retake)/HOA 7.1/WineQualityData/winequality-white.csv', sep = ';')
CData
```

```python
CData.info()
```

```python
CData.corr()
```

```python
CorrMatr = CData.corr()
TargCorr = CorrMatr['quality']
AbsTarCorr = TargCorr.abs()
LowCorrFeat = AbsTarCorr[AbsTarCorr < 0.01]
print(LowCorrFeat)
```

```python
CData = CData.drop(columns = LowCorrFeat.index)
CData
```

```python
CData.info()
```

```python
CX = CData.iloc[:, :-1]
Cy = CData.iloc[:, -1]
```

```python
SS = StandardScaler()
CX = SS.fit_transform(CX)
```

```python
CX_train, CX_test, Cy_train, Cy_test = train_test_split(CX, Cy, test_size = 0.25, random_state = 123)
```

```python
LE = LabelEncoder()
Cy_train = LE.fit_transform(Cy_train)
Cy_test = LE.fit_transform(Cy_test)
Cy_train = to_categorical(Cy_train)
Cy_test = to_categorical(Cy_test)
```

```python
Cy_train.shape
```

##Create a base model

```python
def Cbaseline_model():
  model = Sequential()
  model.add(Dense(32, input_dim = 9, activation = 'relu'))
  model.add(Dense(7, activation = 'softmax'))

  model.compile(loss = 'categorical_crossentropy', optimizer = 'adam', metrics = ['accuracy'])
  return model
```

##Evaluate the model with k-fold cross validation

```python
CEstimator = KerasClassifier(model = Cbaseline_model, epochs = 500, batch_size = 5000, verbose = 0)
Ckfold = KFold(n_splits = 10, shuffle = True)
CResults = cross_val_score(CEstimator, CX_train, Cy_train, cv = Ckfold)
print("Baseline: %.2f%% (%.2f%%)" % (CResults.mean()*100, CResults.std()*100))
```

##Improve the accuracy of your model by applying additional hidden layers

```python
def Cbaseline_model():
  model = Sequential()
  model.add(Dense(128, input_dim = 9, activation = 'relu'))
  model.add(Dense(64, activation = 'relu'))
  model.add(Dense(32, activation = 'relu'))
  model.add(Dense(7, activation = 'softmax'))

  model.compile(loss = 'categorical_crossentropy', optimizer = 'adam', metrics = ['accuracy'])
  return model
```

```python
CEstimator = KerasClassifier(model = Cbaseline_model, epochs = 500, batch_size = 5000, verbose = 0)
Ckfold = KFold(n_splits = 10, shuffle = True)
CResults = cross_val_score(CEstimator, CX_train, Cy_train, cv = Ckfold)
print("Baseline: %.2f%% (%.2f%%)" % (CResults.mean()*100, CResults.std()*100))
```

```python
import pandas as pd
from keras.models import Sequential
from keras.layers import Dense
from scikeras.wrappers import KerasRegressor
from tensorflow.keras.utils import to_categorical
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import KFold
from sklearn.preprocessing import LabelEncoder
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

```python
RData = pd.read_csv('/content/drive/MyDrive/CPE 019 (Retake)/HOA 7.1/WineQualityData/winequality-red.csv', sep = ';')
RData
```

```python
RData.info()
```

```python
RX = RData.iloc[:, :-1]
Ry = RData.iloc[:, -1]
```

```python
#RX.shape
Ry.shape
```

##Create a base model

```python
def Rbaseline_model():
  model = Sequential()
  model.add(Dense(16, input_dim = 11, kernel_initializer = 'normal', activation = 'relu'))
  model.add(Dense(1, activation = 'linear'))

  model.compile(loss = 'mean_squared_error', optimizer = 'adam')
  return model
```

```python
REstimator = KerasRegressor(model = Rbaseline_model, epochs = 100, batch_size = 5000, verbose = 0)
Rkfold = KFold(n_splits = 10)
RResults = cross_val_score(REstimator, RX, Ry, cv = Rkfold, scoring = 'neg_mean_squared_error')
print("Baseline: %.2f (%.2f)" % (RResults.mean(), RResults.std()))
```

##Improve the model by standardizing the dataset

```python
REstimator = []
REstimator.append(('standardize', StandardScaler()))
REstimator.append(('mlp', KerasRegressor(model = Rbaseline_model, epochs = 100, batch_size = 5000, verbose = 0)))
pipeline = Pipeline(REstimator)
Rkfold = KFold(n_splits = 10)
RResults = cross_val_score(pipeline, RX, Ry, cv = Rkfold, scoring = 'neg_mean_squared_error')
print("Baseline: %.2f (%.2f)" % (RResults.mean(), RResults.std()))
```
